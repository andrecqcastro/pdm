```python
# Lendo os dados
reviews_schema = """
marketplace string, 
customer_id long, 
review_id string, 
product_id string, 
product_parent long, 
product_title string, 
product_category string, 
star_rating integer, 
helpful_votes integer, 
total_votes integer, 
vine boolean, 
verified_purchase boolean, 
review_headline string, 
review_body string, 
review_date date
"""

review_toys_df = spark.read.csv(
    "gs://pdm-rabelo-2024/amazon_reviews_us_Toys_v1_00.tsv",
    sep="\t",
    header=True,
    schema=reviews_schema
)

review_toys_df.show(5)
```

```python
# Salvar 'review_toys_df'
review_toys_df.write \
  .format("bigquery") \
  .option("writeMethod", "direct") \
  .save("projeto.review_toys_df")
```
