```python
# Lendo os dados da camada silver
review_toys_df_gold = spark.read \
  .format("bigquery") \
  .load("pdm-class-rabelo-2024.projeto.review_toys_df_silver")

review_toys_df_gold.show(5)
```

```python
# Como estamos preparando nossos dados para rodar inferência, a única coluna que vai nos interessar é a 'review_body'
infer_df = review_toys_df_gold.select("review_body")

# Exibir o novo DataFrame
infer_df.show(5)
```

```python
# Agora uma tabela que será usada para nosso dasboard
dashboard_df = review_toys_df_gold.select(
    "customer_id",
    "review_id",
    "product_id",
    "product_parent",
    "product_title",
    "star_rating",
    "helpful_votes",
    "total_votes",
    "review_headline",
    "review_body",
    "review_date"
)

# Exibir as primeiras linhas do DataFrame para verificar
dashboard_df.show(5)
```

```python
# Salvar 'infer_df'
infer_df.write \
  .format("bigquery") \
  .option("writeMethod", "direct") \
  .save("projeto.infer_df")
```

```python
# Salvar 'dashboard_df'
dashboard_df.write \
  .format("bigquery") \
  .option("writeMethod", "direct") \
  .save("projeto.dashboard_df")
```
